```python
# loading all 10 datasets one by one

files = [
    "01_fund_master.csv",
    "02_nav_history.csv",
    "03_aum_by_fund_house.csv",
    "04_monthly_sip_inflows.csv",
    "05_category_inflows.csv",
    "06_industry_folio_count.csv",
    "07_scheme_performance.csv",
    "08_investor_transactions.csv",
    "09_portfolio_holdings.csv",
    "10_benchmark_indices.csv",
]

for f in files:
    df = pd.read_csv("data/raw/" + f)
    print(f"\n--- {f} ---")
    print("shape  :", df.shape)
    print(df.head(2))
```

```python
# exploring fund master

funds = pd.read_csv("data/raw/01_fund_master.csv")

print("fund houses:", funds["fund_house"].unique())
print("\ncategories:", funds["category"].unique())
print("\nrisk grades:", funds["risk_category"].unique())
```

```python
# checking all codes exist in nav history

nav = pd.read_csv("data/raw/02_nav_history.csv")

fund_codes = set(funds["amfi_code"])
nav_codes = set(nav["amfi_code"])

missing = fund_codes - nav_codes

if len(missing) == 0:
    print("all amfi codes matched!")
else:
    print("missing codes:", missing)
```
